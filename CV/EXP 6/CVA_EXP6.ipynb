```python
import cv2
from google.colab.patches import cv2_imshow # Import cv2_imshow for displaying in Colab

# Load the pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load an image
image_path = "b.png"  # Replace with the actual path of your image
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 3)

# Show the image with detected faces using cv2_imshow
cv2_imshow(img) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import qrcode

# GitHub repository URL
repo_url = "https://github.com/Shree0605/Record/tree/main/CV/EXP%206"

# Create the QR code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,  # Changed to ERROR_CORRECT_L
    box_size=10,
    border=4,
)
qr.add_data(repo_url)
qr.make(fit=True)

# Save the QR code as an image
# Use make_image instead of image to generate the QR code image.
img = qr.make_image(fill_color="black", back_color="white")
img.save("EXP 6.png")
```
